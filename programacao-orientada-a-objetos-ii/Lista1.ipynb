```python
#2 Biblioteca

class Livro():
    def __init__(self, titulo, autor, ano, editora, edicao, volume):
        self.titulo = titulo
        self.autor = autor
        self.ano = ano
        self.editora = editora
        self.edicao = edicao
        self.volume = volume
    
    def getTitulo(self):
        return self.titulo
    def getAutor(self):
        return self.autor
    def getAno(self):
        return self.ano
    def getEditora(self):
        return self.editora
    def getEdicao(self):
        return self.edicao
    def getVolume(self):
        return self.volume


class Biblioteca():
    def __init__(self, livros = []):
        self.livros = livros
    
    def getLivros(self):
        return self.livros
    
    def addLivro(self, livro):
        self.livros.append(livro)

    def removeLivro(self, livro):
        self.livros.remove(livro)

    def buscaTitulo(self, titulo):
        achado = False
        for i in self.livros:
            if i.getTitulo() == titulo:
                achado = True
                return i
        if not achado:
            print("Livro não encontrado.")
    
    def buscaAutor(self, autor):
        achado = False
        for i in self.livros:
            if i.getAutor() == autor:
                achado = True
                return i
        if not achado:
            print("Livro não encontrado.")

    def buscaAno(self, ano):
        achado = False
        for i in self.livros:
            if i.getAno() == ano:
                achado = True
                return i
        if not achado:
            print("Livro não encontrado.")
```

```python
#3 Polinomio

class Polinomio():
    def __init__(self, grau, coeficientes: list):
        self.grau = grau
        self.coeficientes = coeficientes

    def __str__(self):
        poli = ""
        for i in range(self.grau + 1):
            if self.coeficientes[i] < 0:
                poli = poli + f" {self.coeficientes[i]}" + f"x^{i}"
            else:
                poli = poli + f" +{self.coeficientes[i]}" + f"x^{i}"
        return poli.strip()
    
    def __add__(self, other):
        coef_add = []
        coef1 = self.coeficientes
        coef2 = other.coeficientes
        if len(coef1) < len(coef2):
            coef1, coef2 = coef2, coef1
        for i in range(len(coef1)):
            if i < len(coef2):
                coef_add.append(coef1[i] + coef2[i])
            else:
                coef_add.append(coef1[i])
        g_max = max(self.grau, other.grau)
        return Polinomio(g_max, coef_add)
    
    def __mul__(self, other):
        if isinstance(other, Polinomio):
            grau = self.grau + other.grau
            coeficientes = [0 for i in range(grau + 1)]
            for i, valI in enumerate(self.coeficientes):
                for j, valJ in enumerate(other.coeficientes):
                    coeficientes[i + j] += self.coeficientes[i] * other.coeficientes[j]
            return Polinomio(grau, coeficientes)
        
        elif isinstance(other, (int, float)):            
            coeficientes = [i * other for i in self.coeficientes]
            return Polinomio(self.grau, coeficientes)
        
    def calcularEmX(self, valorX):
        resultado = 0
        for i in range(self.grau + 1):
            resultado += valorX**i * self.coeficientes[i]
        return resultado
        
p1 = Polinomio(2, [0, -3, 5])
p2 = Polinomio(3, [0, 4, 3, 1])
print(p1)
print(p1.calcularEmX(2))
print(p1 + p2)
print(p1 * p2)
print(p1* 2)
```

```python
#4 Series

class Serie:
    def __init__(self):
        pass
    
    def fatorial(self, n):
        if n == 1:
            return 1
        return n * self.fatorial(n - 1)
    
    def fibonacci(self, n):
        if n == 0:
            return 0
        elif n == 1:
            return 1
        else:   
            return self.fibonacci(n - 1) + self.fibonacci(n - 2)
            
    def fibonarial(self, n):
        if n == 1:
             return 1
        return self.fibonarial(n - 1) * self.fibonacci(n)

s = Serie()
print(s.fibonacci(7))
print(s.fibonarial(6))
print(s.fatorial(5))
```
